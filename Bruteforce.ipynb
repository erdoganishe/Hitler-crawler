```python
import requests
import time


def get_all_links_from_wiki_page(link):
    
    page = requests.get(link)  #fetch wiki page
    
    content = str(page.content)
    all_a_html = content.split('<a href="') #get all HTML links
    
    #list for 
    hrefs = []
    
    for a_html in all_a_html:
        
        #ignore non-wikipedia.org pages
        #ignore non-article php pages
        #ignore literature links
        #ignore images
        #non-/wiki pages are never articles; ignore
        #ignore for :Special, :Template, etc.
        
        if ("https://" not in a_html \
            and "#" not in a_html \
            and "BookSources" not in a_html \
            and ".jpg" not in a_html \
            and ".jpeg" not in a_html \
            and ".png" not in a_html \
            and ".JPG" not in a_html \
            and ".JPEG" not in a_html \
            and ".PNG" not in a_html \
            and "/w/" not in a_html \
            and ":" not in a_html):
            
            #add child link
            
            link = "https://en.wikipedia.org"
            link += a_html.split('"')[0]
            hrefs.append(link)
    
    return list(set(hrefs))
 
get_wiki_page("https://en.wikipedia.org/wiki/Porajmos")
```

```python
compare_link = "https://en.wikipedia.org/wiki/Oreo"

current_depth = 0

links_db = [{
    'link' : start_link,
    'depth': current_depth,
    'parent': "one"
}]

links_set = [start_link]



def search_for_link(current_depth = 0):
    is_stop = False
    counter = 0
    counter_dubles = 0
    for link_db in links_db:
        if link_db['depth'] == current_depth:
            children = get_all_links_from_wiki_page(link_db['link'])
            
            for child in children: 
                if child == compare_link:
                    is_stop = True
                    links_set.append(child)
                    links_db.append({'link' : child, 'depth': current_depth +1, 'parent': link_db['link']})
                    
                    path = get_path()
                    
                    return path
                
                if child not in links_set:
                    links_set.append(child)
                    links_db.append({'link' : child, 'depth': current_depth +1, 'parent': link_db['link']})
                    counter+=1
                    if counter%1000 == 0:
                        pass
                       # print("Counter ", counter)
                else:
                    counter_dubles += 1
                    if counter_dubles%1000 == 0:
                        pass
                        #print("Counter of dubles is ", counter_dubles)
                    
            
    if is_stop:
        path = get_path()
        return path
    if current_depth < 5:
        current_depth += 1
        #print("Cur_depth is ", current_depth)
        return search_for_link(current_depth)

        
    else:
        return False


def get_first_parent(child):
    for link_db in links_db:
        if link_db['link'] == child:
            return link_db['parent']
        
def get_depth(link):
    for link_db in links_db:
        if link_db['link'] == link:

            return link_db['depth']
        
        
def get_path():
    path = [compare_link]
    depth = get_depth(compare_link)
    current_link = compare_link
    for i in range(0, depth):

        current_link = get_first_parent(current_link)
        path.append(current_link)

        
    return path[::-1]
            
starttime = time.time()



print(search_for_link())
endtime = time.time()
print(endtime-starttime)
```
